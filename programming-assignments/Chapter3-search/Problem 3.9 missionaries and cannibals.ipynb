```python
from collections import deque

def is_valid(M, C):
    """Check if a bank state is safe."""
    if M < 0 or C < 0 or M > 3 or C > 3:
        return False
    return M == 0 or M >= C  # missionaries not outnumbered

def bfs_solution():
    start = (3, 3, 1)   # (M_left, C_left, boat_side)
    goal = (0, 0, 0)
    moves = [(1,0), (2,0), (0,1), (0,2), (1,1)]  # (missionaries, cannibals) crossing

    # BFS queue stores (state, path)
    queue = deque()
    queue.append((start, [start]))
    visited = set()
    visited.add(start)

    while queue:
        (m, c, boat), path = queue.popleft()
        if (m, c, boat) == goal:
            return path

        # Determine which bank people are leaving from
        if boat == 1:   # boat on left → crossing to right
            for dm, dc in moves:
                new_m = m - dm
                new_c = c - dc
                new_boat = 0
                if is_valid(new_m, new_c) and is_valid(3-new_m, 3-new_c):
                    state = (new_m, new_c, new_boat)
                    if state not in visited:
                        visited.add(state)
                        queue.append((state, path + [state]))
        else:            # boat on right → crossing to left
            for dm, dc in moves:
                new_m = m + dm
                new_c = c + dc
                new_boat = 1
                if is_valid(new_m, new_c) and is_valid(3-new_m, 3-new_c):
                    state = (new_m, new_c, new_boat)
                    if state not in visited:
                        visited.add(state)
                        queue.append((state, path + [state]))
    return None

def print_solution(path):
    print("Solution found in", len(path)-1, "moves:")
    for i, (m, c, boat) in enumerate(path):
        side = "left" if boat == 1 else "right"
        print(f"Step {i}: M_left={m}, C_left={c}  boat on {side}")

if __name__ == "__main__":
    sol = bfs_solution()
    if sol:
        print_solution(sol)
    else:
        print("No solution found.")
```
